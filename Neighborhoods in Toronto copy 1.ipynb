```python
    import requests
    import pandas as pd
    from bs4 import BeautifulSoup
    from html.parser import HTMLParser
    import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
website_url = requests.get(url).text
```

```python
soup = BeautifulSoup(website_url,'lxml')
#print(soup.prettify())
```

```python
data = []
table = soup.find('table', attrs={'class':'wikitable sortable'})
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values
```

```python
L=["Postal Code", "Borough", "Neighbourhood"]
```

```python
df=pd.DataFrame(data, columns=L)
df = df.drop([0], axis=0) 
```

```python
dff=df[df.Borough!= 'Not assigned']
```

```python
dff = dff.reset_index()
del dff['index']
```

```python
dff.shape
```

```python
!wget -O Geospatial_Coordinates.csv http://cocl.us/Geospatial_data/Geospatial_Coordinates.csv
```

```python
coordinates = pd.read_csv("Geospatial_Coordinates.csv") 
```

```python
coordinates.head()
```

```python
merged=pd.merge(dff, coordinates, on='Postal Code')
merged
```

```python
def tmp(group_series):
    if (group_series==group_series.iloc[0]).all():
        return group_series.iloc[0]
    else:
        return group_series.sum()
dff2=merged.groupby('Postal Code').agg(tmp)

dff2.reset_index()
```
